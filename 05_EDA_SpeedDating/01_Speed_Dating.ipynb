### Speed Dating Part 1 ###
Start exploring the dataset with pandas to extract some statistics

```python
sd=pd.read_csv('API/src/Speed_Dating_Data.csv', encoding="ISO-8859-1")
pd.set_option('display.max_columns', None)
sd.head(5)
```

```python
print(sd.info(verbose=True))
```

```python
print(sd.columns.tolist())
```

```python
keep_cols = ['gender', 'match', 'int_corr', 'samerace', 'age', 'field', 'field_cd', 'undergra', 'mn_sat', 'tuition', 'race', 'imprace', 'imprelig', 'from', 'zipcode', 'income', 'goal', 'date', 'go_out', 'career', 'career_c', 'sports', 'tvsports', 'exercise', 'dining', 'museums', 'art', 'hiking', 'gaming', 'clubbing', 'reading', 'tv', 'theater', 'movies', 'concerts', 'music', 'shopping', 'yoga', 'exphappy', 'expnum']
keep = [col for col in sd.columns if col in keep_cols]
sd_custom = sd.loc[:,keep]
display(sd_custom.head())
print(f"Number of rows : {sd_custom.shape[0]}")
```

### STATISTICS

```python
print("statistics : ")
sd_custom.describe(include="all")
display(sd_custom.describe(include="all"))
print()
print("Details of dataset with info: ")
display(sd_custom.info())
print()
```

```python
print("Total of null values: ")
print()
sd_null_list = np.array(sd_custom.isnull().sum())
print(sd_null_list)
print()
total = sd.shape[0]
print("Total values: ")
print()
print(total)
print()
print("Percentage of missing values: ")
print()
sd_null_porcentage = ((sd_null_list / total)*100).round(1)
print(sd_null_porcentage)
print()
print()
```

```python
#drop columns with isnull > 40%
sd_clean = sd_custom.loc[:,sd_null_porcentage<=40]
sd_clean.head()
```

### Speed Dating Part 2 ###

```python
#Heatmap for coorelation and selection of columns
corr = sd_clean.corr()
fig, axes = plt.subplots(figsize=(20, 15))
sns.heatmap(corr, annot=True, ax=axes)
plt.show()
```
